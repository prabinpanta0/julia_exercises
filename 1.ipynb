```julia
function is_valid_trangle(a::Real, b::Real, c::Real)::Bool
return (a + b > c) && (a + c > b) && (b + c > a) ? true : false
end

println(is_valid_trangle(4, 5, 10))
println(is_valid_trangle(4, 6, 10))
println(is_valid_trangle(5, 6, 10))

```

```julia
function sum_of_squares(n::Int)::Int
    return (n*(n+1)*(2n + 1) ÷ 6)
end

println(sum_of_squares(4))
println(sum_of_squares(10))
```

```julia
function sum_of_squares(n::Float64)::Float64
    return (n*(n+1)*(2n + 1) / 6)
end

println(sum_of_squares(4.0))
println(sum_of_squares(10.1))
```

```julia
sum_of_squares(n::Int) = (n*(n+1)*(2n + 1) ÷ 6)
sum_of_squares(n::Float64) = (n*(n+1)*(2n + 1) / 6)

println(sum_of_squares(10))
println(sum_of_squares(10.0))
```
